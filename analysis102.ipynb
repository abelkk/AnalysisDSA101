# **Business Understanding**


**About the Dataset**
As Generative AI tools like ChatGPT, Copilot, and Gemini become deeply embedded in academic life, understanding their true impact on students has never been more critical. This dataset provides a rich, multi-dimensional snapshot of student interactions with AI — spanning academic outcomes, behavioral patterns, institutional context, and psychological well-being.

Whether you're a researcher, educator, data scientist, or policymaker, this dataset offers fertile ground for exploring the benefits and risks of AI adoption in higher education.

**Sources**
This dataset was synthetically generated to simulate real-world student behavior and academic outcomes in the context of Generative AI adoption. It is designed to reflect plausible patterns observed in higher education environments where AI tools such as ChatGPT, Microsoft Copilot, and Google Gemini are increasingly used by students across disciplines including STEM, Business, Humanities, Medical, and Arts.

[source](https://www.kaggle.com/datasets/laveshjadon/ai-impact-on-students)


**Collection Methodology**
Data was programmatically generated using statistical distributions and domain-informed rules to ensure realistic relationships between variables. Key design decisions include: - GPA values follow a normal distribution capped at 4.0 - Weekly AI usage hours range from 0 to 40, reflecting light to heavy users - Categorical variables (Major, Policy, Burnout Risk, etc.) are uniformly sampled across meaningful real-world categories - Perceived AI Dependency and Anxiety scores are integer scales (1–10) modeled after standard Likert-type survey instruments - Post-Semester GPA is influenced by Pre-Semester GPA, AI usage, and study hours to simulate realistic academic outcomes - No personally identifiable information (PII) is included; all Student IDs are synthetic The dataset contains 50,000 records with 16 features and zero missing values, making it ready for immediate use in EDA, machine learning, and academic research.
